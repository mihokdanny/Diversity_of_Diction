```python
import re
import numpy as np
import pandas as pd
import nltk
# import requests
# from bs4 import BeautifulSoup
# import urllib.request
```

```python
csv_of_works = "csv_of_works.csv"

list_of_works_df = pd.read_csv(csv_of_works, header=None)
```

```python
list_of_works_df
```

```python
list_of_works = list_of_works_df[0].tolist()
```

```python
list_of_works
```

```python
Test_df = pd.DataFrame(columns=["author_group","author","title","word","frequency","total_words_used","total_non_stop_words","total_unique_non_stop_words"])
```

```python
Test_df
```

```python
for work in list_of_works:
    with open(work + '.txt', 'r') as file:
        text = file.read().replace('\n', ' ')
    tokens = re.findall('\w+', text)
    all_lower = []

    for word in tokens:
        all_lower.append(word.lower())
    total_number_words_used_in_work = len(all_lower)
    sw = nltk.corpus.stopwords.words('english')
    non_stop_words = []

    # Add to words_ns all words that are in words but not in sw
    for word in all_lower:
        if word not in sw:
            non_stop_words.append(word)

    total_non_stop_words = len(non_stop_words)

    dictionary_of_words_used = {}

    for x in non_stop_words:
       if x in dictionary_of_words_used:
           dictionary_of_words_used[x] = dictionary_of_words_used[x] + 1
       else:
           dictionary_of_words_used[x] = 1

    total_unique_non_stop_words_used = len(list(dictionary_of_words_used.keys()))

    dictionary_of_words_used_df = pd.DataFrame(list(dictionary_of_words_used.items()))

    dictionary_of_words_used_df['title'] = work

    dictionary_of_words_used_df = dictionary_of_words_used_df.rename(index=str, columns={0: "word"})

    dictionary_of_words_used_df = dictionary_of_words_used_df.rename(index=str, columns={1: "frequency"})

    dictionary_of_words_used_df['total_words_used'] = total_number_words_used_in_work

    dictionary_of_words_used_df['total_non_stop_words'] = total_non_stop_words

    dictionary_of_words_used_df['total_unique_non_stop_words'] = total_unique_non_stop_words_used

    Test_df = Test_df.append(dictionary_of_words_used_df, ignore_index=True)
              
Test_df.to_csv("Test_df_result.csv", index=False, header=True)
```
